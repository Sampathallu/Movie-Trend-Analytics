```python
from bs4 import BeautifulSoup
import requests
import urllib.request 
import pandas as pd
import numpy as np
import re
import time
```

```python
movies_2018 = pd.DataFrame()
start = 1 
moviename = list()
genre = list()
grossmov = list()
duration = list()
imdbrating = list()
metarating = list()
votes = list()
directormov = list()
starcast = list()
```

```python
for length in range(1,3):
    if (length == 1):
        start = 1
    else:
        start = start + 50

    url_2018 = 'https://www.imdb.com/search/title?title_type=feature&year=2018-01-01,2018-12-31&start='+ str(start) +'&ref_=adv_nxt'
    request = requests.get(url_2018)
    soup = BeautifulSoup(request.content, 'html.parser')
   
    namecontent = soup.find_all('h3', {"class" : 'lister-item-header'})
    for i in namecontent:
    # namecontent = soup.find('h3', {"class" : 'lister-item-header'})
        moviename.append(i.text.strip())
    content = soup.find_all('p', {"class" : 'text-muted '})
    for i in content:
        gen = i.find_all('span', {"class" : 'genre'})
        if (gen):
            for j in gen:
                genre.append(j.text.strip())
        else:
            genre.append("")
    imdbratingcontent = soup.find_all('div', {"class" : 'inline-block ratings-imdb-rating'})
    for i in imdbratingcontent:
        imdbrating.append(i.text.strip())
    divs = soup.findAll('div', {'class':'lister-item'})
    for div in divs:
        spans = div.findAll('span', {'name':'nv'})
        vote = spans[0]['data-value']
        votes.append(vote)
        try:
            gross = spans[1]['data-value']       
        except:
            gross = ""
        grossmov.append(gross)
    duracontent = soup.find_all('p', {"class" : 'text-muted '})
    for i in duracontent:
        gen = i.find_all('span', {"class" : 'runtime'})
        if (gen):
            for j in gen:
                duration.append(j.text.strip())
        else:
            duration.append("0 min")
    ratingsbar = soup.find_all('div', {"class" : 'ratings-bar'})
    for ir in ratingsbar:
        if ir.find_all('div', {"class" : 'inline-block ratings-metascore'}):
            metacontent = ir.find_all('div', {"class" : 'inline-block ratings-metascore'})
            for i in metacontent:
                if (i.find_all('span', {"class" : 'metascore favorable'})):
                    metarating.append(i.text.strip())
                if (i.find_all('span', {"class" : 'metascore mixed'})):
                    metarating.append(i.text.strip())
                if (i.find_all('span', {"class" : 'metascore unfavorable'})):
                    metarating.append(i.text.strip())
        else:
            metarating.append("")
    count = 0
    for directors in soup.find_all(href=re.compile("adv_li_dr")):
        if (bool(re.search('adv_li_dr_0', str(directors))) == 1):
            directormov.append(directors.next_element)
            count = count + 1
        if (bool(re.search('adv_li_dr_1', str(directors))) == 1):
            directormov[count-1] = directormov[count-1] + "," + directors.next_element
    string = ""
    for idx, actors in enumerate(soup.find_all(href=re.compile("adv_li_st"))):
        if (bool(re.search('adv_li_st_0', str(actors))) == 1):
            string = string + actors.next_element
        if (bool(re.search('adv_li_st_1', str(actors))) == 1):
            string = string + "," + str(actors.next_element)
#         if (bool(re.search('adv_li_st_2', str(actors))) == 1):
#             string = string + "," + str(actors.next_element)
            starcast.append(string)
            string = ""
  
   
    time.sleep(5)
```

```python
print(len(moviename))
print(len(genre))
print(len(grossmov))
print(len(duration))
print(len(imdbrating))
print(len(metarating))
print(len(votes))
print(len(directormov))
print(len(starcast))
```

```python
movies_2018['moviename'] = moviename
movies_2018['genre'] = genre
movies_2018['imdbrating'] = imdbrating
movies_2018['votes'] = votes
movies_2018['grossmov'] = grossmov
movies_2018['duration'] = duration
movies_2018['metarating'] = metarating
movies_2018['directormov'] = directormov
movies_2018['starcast'] = starcast
```
