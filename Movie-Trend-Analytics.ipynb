```python
from bs4 import BeautifulSoup
import requests
import urllib.request 
import pandas as pd
import numpy as np
import re
```

```python
url_2018 = 'https://www.imdb.com/search/title?year=2018&title_type=feature&'
```

```python
request = requests.get(url_2018)
request
```

```python
soup = BeautifulSoup(request.content, 'html.parser')
```

```python
soup
```

```python
moviename = list()
genre = list()
grossmov = list()
duration = list()
imdbrating = list()
metarating = list()
votes = list()
directormov = list()
starcast = list()
```

```python
namecontent = soup.find_all('h3', {"class" : 'lister-item-header'})
for i in namecontent:
# namecontent = soup.find('h3', {"class" : 'lister-item-header'})
    moviename.append(i.text.strip())
```

```python
genrecontent = soup.find_all('span', {"class" : 'genre'})
for i in genrecontent:
    genre.append(i.text.strip())
```

```python
imdbratingcontent = soup.find_all('div', {"class" : 'inline-block ratings-imdb-rating'})
for i in imdbratingcontent:
    imdbrating.append(i.text.strip())
```

```python
grossmov = []
votes = []
divs = soup.findAll('div', {'class':'lister-item'})
for div in divs:
    spans = div.findAll('span', {'name':'nv'})
    vote = spans[0]['data-value']
    votes.append(vote)
    try:
        gross = spans[1]['data-value']       
    except:
        gross = ""
    grossmov.append(gross)
```

```python
durationcontent = soup.find_all('span', {"class" : 'runtime'})
for i in durationcontent:
    duration.append(i.text.strip())
```

```python
metacontent = soup.find_all('span', {"class" : 'metascore favorable'})
for i in metacontent:
    metarating.append(i.text.strip())
```

```python
for directors in soup.find_all(href=re.compile("adv_li_dr")):
    director = directors.next_element
    directormov.append(director)
```

```python
for actors in soup.find_all(href=re.compile("adv_li_st")):
    actor = actors.next_element
    starcast.append(actor)
```
