```python
from bs4 import BeautifulSoup
import requests
import urllib.request 
import pandas as pd
import numpy as np
import re
import time
```

```python
movies_2018 = pd.DataFrame()
movies_2017 = pd.DataFrame()
movies_2016 = pd.DataFrame()
start = 1 
moviename = list()
genre = list()
grossmov = list()
duration = list()
imdbrating = list()
metarating = list()
votes = list()
directormov = list()
starcast = list()
releasedate = list()
grosswwtotal = list()
budgettot = list()
imdblinks = list()
```

```python
#2018
for length in range(1,101):
    if (length == 1):
        start = 1
    else:
        start = start + 50

    url_2018 = 'https://www.imdb.com/search/title?title_type=feature&year=2018-01-01,2018-12-31&start='+ 'start' +'&ref_=adv_nxt'
    request = requests.get(url_2018)
    soup = BeautifulSoup(request.content, 'html.parser')
   
    star = soup.find_all('div', {"class" : 'lister-item-content'})
    for stars in star:
        st = stars.find_all(href=re.compile("adv_li_tt"))
        moviename.append(st[0].text.strip())
    content = soup.find_all('p', {"class" : 'text-muted '})
    for i in content:
        gen = i.find_all('span', {"class" : 'genre'})
        if (gen):
            for j in gen:
                genre.append(j.text.strip())
        else:
            genre.append("")
    star = soup.find_all('div', {"class" : 'lister-item-content'})
    for ist in star:
        if (ist.find_all('div', {"class" : 'ratings-bar'})):
            ratingsbar = ist.find_all('div', {"class" : 'ratings-bar'})
            for ir in ratingsbar:
                if ir.find_all('div', {"class" : 'inline-block ratings-imdb-rating'}):
                    mcontent = ir.find_all('div', {"class" : 'inline-block ratings-imdb-rating'})
                    for i in mcontent:
                        if (i.find_all('span', {"class" : 'global-sprite rating-star imdb-rating'})):
                            imdbrating.append(i.text.strip())
                else:
                    imdbrating.append('0')
        else:
            imdbrating.append('0')
    divs = soup.findAll('div', {'class':'lister-item'})
    for div in divs:
        spans = div.findAll('span', {'name':'nv'})
        try: 
            vote = spans[0]['data-value']
            gross = spans[1]['data-value']  
        except:
            vote = 0
            gross = ""
        votes.append(vote)
        grossmov.append(gross)
    duracontent = soup.find_all('p', {"class" : 'text-muted '})
    for i in duracontent:
        gen = i.find_all('span', {"class" : 'runtime'})
        if (gen):
            for j in gen:
                duration.append(j.text.strip())
        else:
            duration.append("0 min")
    meta = soup.find_all('div', {"class" : 'lister-item-content'})
    for ism in meta:
        if (ism.find_all('div', {"class" : 'ratings-bar'})):
            ratingsbar = ism.find_all('div', {"class" : 'ratings-bar'})
            for ir in ratingsbar:
                if ir.find_all('div', {"class" : 'inline-block ratings-metascore'}):
                    metacontent = ir.find_all('div', {"class" : 'inline-block ratings-metascore'})
                    for i in metacontent:
                        if (i.find_all('span', {"class" : 'metascore favorable'})):
                            k = i.find('span', {"class" : 'metascore favorable'})
                            metarating.append(k.text.strip())
                        if (i.find_all('span', {"class" : 'metascore mixed'})):
                            l = i.find('span', {"class" : 'metascore mixed'})
                            metarating.append(l.text.strip())
                        if (i.find_all('span', {"class" : 'metascore unfavorable'})):
                            m = i.find('span', {"class" : 'metascore unfavorable'})
                            metarating.append(m.text.strip())
                else:
                    metarating.append('0')
        else:
            metarating.append('0')
    bar = soup.find_all('div', {"class" : 'lister-item-content'})
    for irs in bar:
        if irs.find_all(href=re.compile("adv_li_dr")):
            z = irs.find_all(href=re.compile("adv_li_dr_0"))
            directormov.append(z[0].text.strip())
        else:
            directormov.append("")
    string = ""
    star = soup.find_all('div', {"class" : 'lister-item-content'})
    for stars in star:
        if stars.find_all(href=re.compile("adv_li_st")):
            st = stars.find_all(href=re.compile("adv_li_st"))
            if (bool(re.search('adv_li_st_0', str(st))) == 1):
                string = string + st[0].text.strip()
                if (bool(re.search('adv_li_st_1', str(st))) == 1):
                    string = string + "," + str(st[1].text.strip())
                    starcast.append(string)
                    string = ""
                else:
                    starcast.append(string)
                
        else:
            starcast.append("")
 
    starss = soup.find_all('div', {"class" : 'lister-item-content'})
    for stars in starss:
        st = stars.find_all(href=re.compile("adv_li_tt"))
        link = st[0]['href']
        imdblink = "https://www.imdb.com/" + link
        imdblinks.append(imdblink)
#     for idx,i in enumerate(imdblinks):
#         request1 = requests.get(i)
#         soup1 = BeautifulSoup(request1.content, 'html.parser')
#         grossdetails = soup1.find_all('div', {"id" : 'titleDetails'})
#         for gsd in grossdetails:
#             gsds = gsd.find_all('h4', {"class" : 'inline'})
#             for j in gsds:
#                 if (j.text.strip() == 'Cumulative Worldwide Gross:'):
#                     byt = j.nextSibling.encode('utf-8')
#                     grossworldwide[idx] = "".join( chr(x) for x in byt)
#                 if (j.text.strip() == 'Budget:'):
#                     byt = j.nextSibling.encode('utf-8')
#                     budget[idx] = "".join( chr(x) for x in byt)
#         reldat = soup1.find_all('div', {"class" : 'titleBar'})
#         for rdt in reldat:
#             rd = rdt.find('a', {"title" : 'See more release dates'})
#             releasedate.append(rd.text.strip())  

    time.sleep(0.1)
grossworldwide = [''] * len(imdblinks)
budget = [''] * len(imdblinks)
```

```python
# star = soup.find_all('div', {"class" : 'lister-item-content'})
for idx,i in enumerate(imdblinks):
    request1 = requests.get(i)
    soup1 = BeautifulSoup(request1.content, 'html.parser')
    grossdetails = soup1.find_all('div', {"id" : 'titleDetails'})
    for gsd in grossdetails:
        gsds = gsd.find_all('h4', {"class" : 'inline'})
        for j in gsds:
            if (j.text.strip() == 'Cumulative Worldwide Gross:'):
                byt = j.nextSibling.encode('utf-8')
                grossworldwide[idx] = "".join( chr(x) for x in byt)
            if (j.text.strip() == 'Budget:'):
                byt = j.nextSibling.encode('utf-8')
                budget[idx] = "".join( chr(x) for x in byt)

    reldat = soup1.find_all('div', {"class" : 'titleBar'})
    for rdt in reldat:
        rd = rdt.find('a', {"title" : 'See more release dates'})
        if (rd):
            releasedate.append(rd.text.strip())   
        else:
            releasedate.append("")   
# grosswwtotal.extend(grossworldwide)
# budgettot.extend(budget)
```

```python
movies_2018['moviename'] = moviename
movies_2018['genre'] = genre
movies_2018['imdbrating'] = imdbrating
movies_2018['votes'] = votes
movies_2018['grossmov'] = grossmov
movies_2018['duration'] = duration
movies_2018['metarating'] = metarating
movies_2018['directormov'] = directormov
movies_2018['starcast'] = starcast
movies_2018['releasedate'] = releasedate
movies_2018['grossworldwide'] = grossworldwide
movies_2018['budget'] = budget
```
